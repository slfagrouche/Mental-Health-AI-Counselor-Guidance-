```python
# Mental Health Counselor System – EDA Notebook

import os
import re
import json
import math
from pathlib import Path

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from collections import Counter

from textblob import TextBlob
from wordcloud import WordCloud
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sentence_transformers import SentenceTransformer
import lightgbm as lgb
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from scipy.stats import uniform, randint
import numpy as np

import kagglehub
from kagglehub import KaggleDatasetAdapter


file_path = "train.csv"

df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "thedevastator/nlp-mental-health-conversations",
  file_path,
)


print("First 5 records:", df.head())


# Ensure NLTK resources are available
import nltk
for resource in ["punkt", "wordnet", "stopwords"]:
    try:
        nltk.data.find(f"tokenizers/{resource}")
    except LookupError:
        nltk.download(resource)

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
nltk.download('punkt_tab')

print("Dataset Shape:", df.shape)
df.head()

df.info()


print("\nMissing Values (per column):\n", df.isna().sum())

print("Duplicate rows:", df.duplicated().sum())


# Data Loading & Initial Examination (Additions)
print(f"\nShape before cleaning: {df.shape}")

# Handle missing responses
df.dropna(subset=['Response'], inplace=True)
print(f"Shape after dropping NA in Response: {df.shape}")

# Handle duplicate conversation pairs
df.drop_duplicates(inplace=True)
print(f"Shape after dropping duplicates: {df.shape}")

# Reset index after dropping rows
df.reset_index(drop=True, inplace=True)



# Keep a clean copy for later reference
raw_df = df.copy()

#  Text Pre‑processing Helpers
STOPWORDS = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def clean_text(text: str) -> str:
    """Lowercase, remove non‑alphabetic, tokenize, remove stopwords, lemmatize."""
    tokens = word_tokenize(re.sub(r"[^a-zA-Z']", " ", str(text).lower()))
    tokens = [lemmatizer.lemmatize(tok) for tok in tokens if tok not in STOPWORDS and len(tok) > 2]
    return " ".join(tokens)

df["context_clean"] = df["Context"].apply(clean_text)
df["response_clean"] = df["Response"].apply(clean_text)

# 3.5 Generate Sentence Embeddings
print("Loading Sentence Transformer model...")
embedder = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded.")

# Generate embeddings
print("Generating context embeddings...")
df['context_embedding'] = df['Context'].apply(lambda x: embedder.encode(str(x))) # Use original context for richer embeddings
print("Generating response embeddings...")
df['response_embedding'] = df['Response'].apply(lambda x: embedder.encode(str(x))) # Use original response

print("Embeddings generated.")
embedding_dim = len(df['context_embedding'][0])
print(f"Embedding dimension: {embedding_dim}")



#  Message Length & Structure Analysis

df["context_len"] = df["context_clean"].str.split().apply(len)
df["response_len"] = df["response_clean"].str.split().apply(len)

fig, ax = plt.subplots(1, 2, figsize=(14, 5))
ax[0].hist(df.context_len, bins=50)
ax[0].set_title("Patient Message Length (Token Count)")
ax[1].hist(df.response_len, bins=50)
ax[1].set_title("Psychologist Response Length (Token Count)")
plt.show()

print(df[["context_len", "response_len"]].describe(percentiles=[.25, .5, .75, .9]))

# Lexical & N‑gram Exploration

def top_ngrams(corpus, ngram_range=(1, 1), top_n=20):
    vec = TfidfVectorizer(stop_words="english", ngram_range=ngram_range)
    X = vec.fit_transform(corpus)
    sum_tfidf = X.sum(axis=0)
    tfidf_scores = [(word, sum_tfidf[0, idx]) for word, idx in vec.vocabulary_.items()]
    return sorted(tfidf_scores, key=lambda x: x[1], reverse=True)[:top_n]

print("Top unigrams (Patient Context):")
print(top_ngrams(df.context_clean, (1, 1)))

print("Top bigrams (Patient Context):")
print(top_ngrams(df.context_clean, (2, 2)))

# WordCloud for psychologist responses
wordcloud_text = " ".join(df.response_clean)
WordCloud(width=800, height=400, background_color="white").generate(wordcloud_text).to_image().show()

#  Sentiment & Emotion Analysis


df["context_polarity"] = df['Context'].apply(lambda t: TextBlob(str(t)).sentiment.polarity)
df["response_polarity"] = df['Response'].apply(lambda t: TextBlob(str(t)).sentiment.polarity)


plt.hist(df.context_polarity, bins=50, alpha=0.6, label="Patient")
plt.hist(df.response_polarity, bins=50, alpha=0.6, label="Psychologist")
plt.title("Sentiment Polarity Distributions")
plt.legend()
plt.show()

print("Average polarity (Context, Response):",
      df.context_polarity.mean(), df.response_polarity.mean())

#  Topic Modeling (TF‑IDF + LDA)

vectorizer = TfidfVectorizer(max_features=5000, stop_words="english")
X = vectorizer.fit_transform(df.context_clean)

lda = LatentDirichletAllocation(n_components=10, random_state=42, learning_method="batch")
lda_topics = lda.fit_transform(X)

def display_topics(model, feature_names, no_top_words):
    for idx, topic in enumerate(model.components_):
        print(f"\nTopic {idx}:")
        print("  " + ", ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

display_topics(lda, vectorizer.get_feature_names_out(), 10)

df[[f"topic_{i}" for i in range(10)]] = lda_topics

df["dominant_topic"] = df[[f"topic_{i}" for i in range(10)]].idxmax(axis=1)

plt.bar(df.dominant_topic.value_counts().index, df.dominant_topic.value_counts())
plt.title("Distribution of Dominant Topics in Patient Messages")
plt.xlabel("Topic")
plt.ylabel("Count")
plt.show()

# Response Pattern Analysis

QUESTION_REGEX = re.compile(r"\?$")

def categorize_response(text: str) -> str:
    """Lowercase, check for specific phrases to categorize response type."""
    try:  # Attempt to convert to lowercase
        lower = text.lower()
    except AttributeError:  # Handle non-string values (like floats)
        lower = str(text).lower()  # Convert to string before lowercase
    if "i understand" in lower or "that sounds" in lower:
        return "Validation"
    elif QUESTION_REGEX.search(str(text).strip()):  # Convert 'text' to string before strip
        return "Question"
    elif any(x in lower for x in ["should", "could", "try", "recommend"]):
        return "Advice"
    else:
        return "Empathetic Listening"

# Apply and inspect distribution

df["response_type"] = df["Response"].apply(categorize_response) # Changed df.response to df['Response'] to match the actual column name.
print(df.response_type.value_counts(normalize=True) * 100)

sns.countplot(y=df.response_type)
plt.title("Psychologist Response Type Distribution (Heuristic)")
plt.show()

# Correlation with dominant patient topic
pd.crosstab(df.dominant_topic, df.response_type, normalize="index").plot(kind="bar", stacked=True)
plt.title("Response Type vs. Patient Topic")
plt.ylabel("Proportion")
plt.show()

# 9 Revised Feature Engineering (Embeddings + Numeric)

numeric_features = [
    "context_len",
    "context_polarity",
] + [f"topic_{i}" for i in range(10)] # LDA topics from context

# Embedding column names
embedding_context_col = 'context_embedding'
embedding_response_col = 'response_embedding'

# Target variable
target = 'response_type'

print("Numeric Features:", numeric_features)
print("Embedding Columns:", embedding_context_col, embedding_response_col)
print("Target:", target)

# Prepare X (containing columns names) and y
X = df[numeric_features + [embedding_context_col, embedding_response_col]]
y = df[target]

# Advanced Classifier: LightGBM with Embeddings & Tuning


# Split data first (contains numeric columns + columns with embedding arrays)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

print(f"Training set shape (before processing): {X_train.shape}")
print(f"Test set shape (before processing): {X_test.shape}")

# --- Feature Processing ---
# 1. Scale Numeric Features
scaler = StandardScaler()
X_train_numeric_scaled = scaler.fit_transform(X_train[numeric_features])
X_test_numeric_scaled = scaler.transform(X_test[numeric_features])

# 2. Extract and Stack Embeddings
# Convert embedding columns to dense numpy arrays
X_train_embed_context = np.array(X_train[embedding_context_col].tolist())
X_test_embed_context = np.array(X_test[embedding_context_col].tolist())
X_train_embed_response = np.array(X_train[embedding_response_col].tolist())
X_test_embed_response = np.array(X_test[embedding_response_col].tolist())

# 3. Combine Scaled Numeric + Embeddings
X_train_processed = np.hstack((X_train_numeric_scaled, X_train_embed_context, X_train_embed_response))
X_test_processed = np.hstack((X_test_numeric_scaled, X_test_embed_context, X_test_embed_response))

print(f"Processed training features shape: {X_train_processed.shape}")
print(f"Processed test features shape: {X_test_processed.shape}")

# --- LightGBM Model and Tuning ---
# Define the parameter distribution for Randomized Search
param_dist = {
    'n_estimators': randint(100, 1000),
    'learning_rate': uniform(0.01, 0.2),
    'num_leaves': randint(20, 60),
    'max_depth': randint(5, 20),
    'feature_fraction': uniform(0.6, 0.4), # range [0.6, 1.0]
    'bagging_fraction': uniform(0.6, 0.4), # range [0.6, 1.0]
    'bagging_freq': randint(1, 10),
    'min_child_samples': randint(10, 50),
    'reg_alpha': uniform(0, 1), # L1 regularization
    'reg_lambda': uniform(0, 1), # L2 regularization
}

# Initialize LightGBM Classifier
# Using is_unbalance=True for imbalance handling
lgbm = lgb.LGBMClassifier(objective='multiclass',
                          metric='multi_logloss', # Logloss often used, but optimize for f1_macro
                          is_unbalance=True, # Alternative to scale_pos_weight
                          random_state=42,
                          n_jobs=-1) # Use all available cores

# Setup Randomized Search with Cross-Validation
n_iter_search = 50 # Number of parameter settings that are sampled. Increase for better search.
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

random_search = RandomizedSearchCV(lgbm,
                                   param_distributions=param_dist,
                                   n_iter=n_iter_search,
                                   cv=cv_strategy,
                                   scoring='f1_macro', # Optimize for macro F1-score
                                   n_jobs=-1, # Use all cores for CV folds
                                   verbose=1, # Show progress
                                   random_state=42)

print(f"\nStarting Randomized Search CV (n_iter={n_iter_search}, cv=5)...")
random_search.fit(X_train_processed, y_train)
print("Randomized Search complete.")

# --- Evaluation ---
print(f"\nBest parameters found by Randomized Search:")
print(random_search.best_params_)
print(f"\nBest cross-validation F1 Macro score: {random_search.best_score_:.4f}")

# Get the best estimator
best_lgbm = random_search.best_estimator_

# Predict on the test set using the best estimator
print("\nEvaluating the best model on the test set...")
preds = best_lgbm.predict(X_test_processed)

# Evaluate the best model
print("\nClassification Report (Best Tuned LightGBM):")
print(classification_report(y_test, preds))

print("Confusion Matrix (Best Tuned LightGBM):\n")
cm = confusion_matrix(y_test, preds, labels=best_lgbm.classes_)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=best_lgbm.classes_, yticklabels=best_lgbm.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix (Best Tuned LightGBM)')
plt.show()
```
